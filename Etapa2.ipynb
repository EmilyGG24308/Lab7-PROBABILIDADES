## Etapa 2: Análisis de la probabilidad de éxito en función del número de sobres

- Número total de estampas: \(N = 100\)
- Estampas por sobre: \(S = 7\)
- Simulaciones: \(R = 10000\)

Se analizará la probabilidad de completar el álbum comprando una cantidad fija de sobres.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(2026)

sns.set_theme(style="whitegrid")
```

```python
N = 100
S = 7
R = 10000
```

## Función de simulación

```python
def simular_M_sobres(M, N=100, S=7):

    album = np.zeros(N, dtype=bool)

    for _ in range(M):

        sobre = np.random.choice(
            N,
            size=S,
            replace=False
        )

        album[sobre] = True

    return album.all()
```

## Simulación Monte Carlo

```python
import seaborn as sns
```

```python

# Valores de M a evaluar
M_values = [20, 25, 30, 35, 40, 45, 50, 60, 70, 80]

# Lista para guardar probabilidades
probabilidades_exito = []
```

```python
# Simulación para cada cantidad de sobres
for M in M_values:

    exitos = 0

    for _ in range(R):

        if simular_M_sobres(M, N=N, S=S):
            exitos += 1

    probabilidad = exitos / R

    probabilidades_exito.append(probabilidad)
```

```python
print("RESULTADOS ETAPA 2 \n")

for M, p in zip(M_values, probabilidades_exito):

    print(f"M = {M:2d} sobres --> P(completar álbum) = {p:.4f}")
```

### Visualización de resultados

La siguiente gráfica muestra cómo aumenta la probabilidad de completar el álbum conforme aumenta el número de sobres comprados.

```python
plt.figure(figsize=(10, 6))

sns.barplot(
    x=M_values,
    y=probabilidades_exito,
    color="cornflowerblue"
)

plt.axhline(
    y=0.5,
    color="red",
    linestyle="--",
    linewidth=2,
    label="Probabilidad = 0.5"
)

plt.title("Probabilidad de completar el álbum vs número de sobres")

plt.xlabel("Número de sobres comprados (M)")

plt.ylabel("Probabilidad estimada")

plt.ylim(0, 1)

plt.legend()

plt.show()
```

#### 2. Comparar el valor de \(M\) que da una probabilidad del 50% con la mediana del número de sobres necesarios.

El valor de \(M\) donde:

\[
P(T \le M) \approx 0.5
\]

corresponde a la definición de la mediana de la variable aleatoria \(T\), donde \(T\) representa el número de sobres necesarios para completar el álbum.

Por ello, el valor obtenido experimentalmente para una probabilidad cercana al 50% debe ser similar a la mediana observada en la distribución de la Etapa 1.

#### 3. Cota de la Unión (Boole) para \(M = 50\)

La probabilidad aproximada de que una estampa específica no aparezca después de comprar \(M\) sobres es:

\[
P(\text{falta una estampa}) \approx e^{-MS/N}
\]

Aplicando la desigualdad de Boole:

\[
P(\text{falta al menos una estampa})
\le
N e^{-MS/N}
\]

Sustituyendo:

\[
N = 100, \quad S = 7, \quad M = 50
\]

se obtiene:

\[
P(\text{fallo})
\le
100 e^{-50(7)/100}
\]

\[
=
100 e^{-3.5}
\]

\[
\approx
100(0.0302)
\]

\[
\approx 3.02
\]

Como una probabilidad no puede ser mayor que 1, la cota se vuelve trivial:

\[
P(\text{fallo}) \le 1
\]

Por ello, la cota de Boole no resulta útil en este caso, ya que sobreestima considerablemente la probabilidad de fallo.
